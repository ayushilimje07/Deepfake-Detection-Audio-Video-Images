```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np
import os
import glob
import cv2
import pandas as pd
from tqdm import tqdm
```

```python
# Load the pre-trained ResNet50 model for feature extraction
resnet_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')
```

```python
# Define function to extract features from face-cropped videos
def extract_features_from_video(video_path):
    cap = cv2.VideoCapture(video_path)
    frames_features = []
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Preprocess frame for ResNet50 (224x224 input size)
        resized_frame = cv2.resize(frame, (224, 224))
        img_array = np.expand_dims(resized_frame, axis=0)
        img_array = preprocess_input(img_array)

        # Extract features from ResNet50
        features = resnet_model.predict(img_array)
        frames_features.append(features)

    cap.release()
    return np.array(frames_features)
```

```python
# Save extracted features for each video
def save_features(video_files, out_dir, label):
    # Ensure the output directory exists
    label_dir = os.path.join(out_dir, label)
    os.makedirs(label_dir, exist_ok=True)

    for video_path in tqdm(video_files):
        features = extract_features_from_video(video_path)
        video_name = os.path.basename(video_path).split('.')[0]
        
        # Save features in the appropriate subdirectory
        np.save(os.path.join(label_dir, f"{video_name}_features.npy"), features)
```

```python
# Load metadata and extract features
def process_metadata_and_extract_features(metadata_path, videos_dir, output_features_base_dir):
    # Read the metadata file
    metadata = pd.read_csv(metadata_path)
    
    # Iterate through each row in the metadata
    for index, row in metadata.iterrows():
        video_name = row[0]  # Assuming the video name is in the first column
        label = row[1]       # Assuming the label is in the second column
        
        # Construct the full path to the video
        video_path = os.path.join(videos_dir, video_name)
        
        # Check if the video exists
        if os.path.exists(video_path):
            print(f"Processing video: {video_name} with label: {label}")
            save_features([video_path], output_features_base_dir, label)
        else:
            print(f"Video {video_name} does not exist at {video_path}")
```
